```python
import requests
import bs4
content = requests.get("https://www.espncricinfo.com/records/tournament/team-match-results/indian-premier-league-2023-15129")
data = content.text
soup = bs4.BeautifulSoup(data)
tables = soup.find_all('table')
anchor_list = soup.findAll("a",{'title':'Twenty20'})

hrefs=[]
website = "https://www.espncricinfo.com"
for i in anchor_list:
    link = website + i.get("href")
    hrefs.append(link)
matches =[]
match_no=0
for i in range(len(hrefs)):
    content = requests.get(hrefs[i])
    data = content.text
    soup = bs4.BeautifulSoup(data)
    title= soup.find('title')
    match_title = title.text.split(",",1)
    tables = soup.findAll("table")
    match = []
    for j in range(len(tables)):
        rows=[]
        header=[]
        trs = tables[j].findAll("tr")
        ths = tables[j].findAll("th")

        for th in ths:
            header.append(th.text) 
        header.append("Match")
        header.append("Match Number")

        
        df = pd.DataFrame(columns=header)
        flag = 0
        for tr in trs:
            row=[]
            tds = tr.findAll("td")   
            for td in tds:
                row.append(td.text)
            
            
            if (j%2==0):
                if (len(row) == 8):
                    row.append(match_title[0])
                    row.append(match_no)
                    df.loc[len(df)] = row
            else:
                if(len(row) == 11):
                    row.append(match_title[0])
                    row.append(match_no)
                    df.loc[len(df)] = row
        if len(df)!=0:
            match.append(df.to_json())
    matches.append(match)
    print(match_no)
    match_no+=1
```

```python
batting_summary=[]
bowling_summary=[]
for i in range(len(matches)):
    for j in range(len(matches[i])):
        if j%2 == 0:
            batting_summary.append(matches[i][j])
        else:
            bowling_summary.append(matches[i][j])
```

```python
batting_summary
```

```python
batting_df = pd.concat([pd.read_json(batting_summary[i]) for i in range(len(batting_summary))],ignore_index= False )
```

```python
bolwing_df =  pd.concat([pd.read_json(bowling_summary[i]) for i in range(len(bowling_summary))],ignore_index= False )
```

```python
batting_df.to_excel("Batting_IPL2023.xlsx")
bolwing_df.to_excel("Bowling_IPL2023.xlsx")
```

```python
import xlsxwriter
f=0
for href in hrefs:
    content = requests.get(href)
    data = content.text
    soup = bs4.BeautifulSoup(data)
    details = soup.findAll("div",{'class':'ds-relative ds-flex ds-flex-row ds-space-x-4 ds-p-3'})
    players=[]
    print(href)
    
    for i in range(len(details)):
        player_info = details[i].findAll("span")
        temp = []
        for j in range(len(player_info)):
            if(len(player_info)>=9 and j==2 or player_info[j].text=='Withdrawn'):
                continue  
            temp.append(player_info[j].text)
        temp.append(team[f])
        players.append(temp) 
    df = pd.DataFrame(players)
    df_list.append(df)
    f+=1
    
```
